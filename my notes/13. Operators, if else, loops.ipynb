```python
# take one 3 digit input from user
number=int(input("Enter a 3 digit number: ")) # 3 digit no stored in number
# take each digits one by one anyhow and store them in variable

a=number%10   # extracting last digit from the 3 digit and stored in a...if we divide any no. by 10 then remainder will be the last no.
number=number//10  #  now 2 digit stored in the number
b=number%10     #extracting last digit from the 2 digit and stored in b
number=number//10  # now 1 digit stored in the last number
c=number%10       #extracting last digit from this 1 digit

# print result
print("result: ", a+b+c)
```

```python
fnum=int(input("Enter first number: "))
snum=int(input("Enter second number: "))

op=input("Enter the operator: ")

if op=="+":
  print(fnum+snum)
elif op=="-":
  print(fnum-snum)
elif op=="*":
  print(fnum*snum)
elif op=="/":
  print(fnum/snum)
else:
  print("Invalid operator")
```
