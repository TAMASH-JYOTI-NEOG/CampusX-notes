# Condition to use For and while loop:  
* Use a for loop when you know the number of iterations in advance, such as when iterating over a sequence (e.g., list, range, or string).
* Use a while loop when the iterations depend on a dynamic condition that continues until it becomes false.




## *`Sequence sum`*

*   1/1! + 2/2! + 3/3! ...





```python
# start code here
n=int(input("Enter n: "))

result=0
fact=1

for i in range(1,n+1):
  fact=fact*i
  result=result + i/fact

print(result)
```

scenerios where loops under a loop is need that are nested loop concept.
* For example if we needed to print all students name of all classes of a school---first we will make a loop to print all classes names from 1 to 10----then under this we will make another loop to print all student name of the particular class.


```python
# unique pairs of a range of nos----36 unique pairs of a 1-6 range of a dice

for i in range(1,7):                # so nested loop eise kaam ata hei jab hamei uper ek chiz chalana hei and
  for j in range(1,7):              # and then uske under hamei koi bhi chiz bar bar karna hei
    print(i,j)                      # like ek classroom mei gaya waha sabhi students ka naam lia
                                    # then another class mei gaya and sabhi students ka naam kia and so on....
```

# **Loops Control Statement**
* **Break** - The break statement is used to exit a loop immediately when a specific condition is met.  
Real life example --- searching is an eg., let whenever i am seraching a particular name in a database of 1 crore then its impossible to find that particular name without using the break
* **Continue** - The continue statement is used to skip the current iteration of a loop and move to the next iteration, without exiting the loop.  
Real life example --- suppose in any ecommerce site i want display all products but dont want to display or skip out of stock products, then use continue to skip out of stock products.
* **Pass** - The pass statement is used to skip execution in a block where code is required but isn't written yet. But will write the code in future where the pass is placed. Just error se baasne ka ek mechnism hei kyunki agar us indended block mei kuch coding nehi karoge toh error he dega

```python
#loop break concept example
for i in range(1,10):
  if i==5:                          #if in the range(1,10), i==5 mil jae toh loop break kardo, yani iske baad loop kam nehi karega
    break
  print(i)
```

## **`Q. loop break question - `**  
Make a program take inputs from user for a range and print all prime nos in that user input range.


```python
lower=int(input("Enter lower range: "))
upper=int(input("Enter upper range: "))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i)
```

```python
#loop continue concept example
for i in range(1,10):
  if i==5:                       #if in the range(1,10), i==5 mil bhi jae toh ise skeep karo and rest loop continue karo
    continue
  print(i)
```

## 2. Accessing strings  
* `Indexing`-- to extract any index characters from a string



```python
# positive indexing----python assigns index to each characters of the string from beginning from 0,1...
# we can extract characters by using index position in [] like s[0]

s="hello world"
print(s[1])
```

```python
# negative indexing---python assigns index to each characters of the string from last from -1,-2...
# benefit of -ve indexing----without even knowing the length of the string we can very easily extract last chars
##otherwise if use dont had -ve index i have to known the length of the string to extract last chars by +ve indexing

s="hello world"
print(s[-1])
```

* **`Slicing`**---To extract multiple characters/substrings from a string by give them a range like s[0:6]
* **`Step Size`** is also used with slicing which give us more flexibility to extract according to me

```python
# slicing with by default +1 step size

s='hello world'

print(s[0:5])
print(s[2:])
print(s[:3])
print(s[:])
```

```python
# slicing with +ve step size----if want to use +ve step size, then range must be lowest to highest

s='hello world'

print(s[0:5:3])
```

```python
# slicing with -ve step size----if want to use -ve step size, then range must be highest to lowest

s='hello world'

print(s[-1:-6:-1])
print(s[6:0:-1])
print(s[:0:-1])
print(s[100:0:-1])
print(s[::-1])               #famous to reverse the full string
```
