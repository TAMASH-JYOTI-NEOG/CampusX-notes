### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.---------distance=square root of (x2-x1)^2 + (y2-y1)^2


> Add blockquote



````python
# Write your code here
p1x=int(input("Enter point 1's x coordinate: "))
p1y=int(input("Enter point 1's y coordinate: "))
p2x=int(input("Enter point 2's x coordinate: "))
p2y=int(input("Enter point 2's y coordinate: "))

euclidian_distance=((p2x-p1x)**2 + (p2y-p1y)**2)**0.5

print(round(euclidian_distance,2))   #round fn give a large no in round figure...here using 2 point round fig
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

*hint---an=a+(n-1)d

````python
# Write your code here
first_term=int(input("Enter first term: "))         #first=1
second_term=int(input("Enter second term: "))       #second=3
n=int(input("Enter the value of n: "))              #n=8

d=second_term-first_term                            #d=3-1=2

an=first_term+(n-1)*d         #1,3,5,7,9,11,13,15------15 is the 8th term in the sequence
print(an)                                           # Result= 15
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.

**hint**---milk tank is like a cube, glasses are like cylinder
====milk tank volume/1glass milk---can solve our problem


Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
l_tank=float(input("Enter the length of the tank: "))
b_tank=float(input("Enter the breadth of the tank: "))
h_tank=float(input("Enter the height of the tank: "))

r_glass=float(input("Enter the radius of the glass: "))
h_glass=float(input("Enter the height of the glass: "))

vol_tank=l_tank*b_tank*h_tank                      # vol of a cube=lbh
vol_glass=3.14*r_glass*r_glass*h_glass             # vol of a cylinder=pie*r2*h

print("no. of glasses are: ",vol_tank/vol_glass)

#below we again writing the code but by use math.floor module fn to approximate the decimals to a complete no.
````

````python
# Write your code here
import math          # just to use floor funtion in the print

l_tank=float(input("Enter the length of the tank: "))
b_tank=float(input("Enter the breadth of the tank: "))
h_tank=float(input("Enter the height of the tank: "))

r_glass=float(input("Enter the radius of the glass: "))
h_glass=float(input("Enter the height of the glass: "))

vol_tank=l_tank*b_tank*h_tank                      # vol of a cube=lbh
vol_glass=3.14*r_glass*r_glass*h_glass             # vol of a cylinder=pie*r2*h

print("no. of glasses are: ",math.floor(vol_tank/vol_glass))        #math.floor use to approximate the decimals to a complete no.
````
